```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv('data-6-02-25.csv')
```

```python
# Extract keywords and count them
keywords = df['Keywords'].dropna().str.split(',').sum()
keyword_counts = Counter(keywords)
# Filter out keywords that are used less than once
filtered_keyword_counts = {k: v for k, v in keyword_counts.items() if v > 1}
sorted_keyword_counts = dict(sorted(filtered_keyword_counts.items(), key=lambda item: item[1], reverse=False))

print(keywords)
```

```python
plt.figure(figsize=(10, 8))
# plt.pie(sorted_keyword_counts.valuesv(), labels=sorted_keyword_counts.keys(), autopct='%1.1f%%', startangle=90)
plt.pie(filtered_keyword_counts.values(), labels=filtered_keyword_counts.keys(), autopct='%1.1f%%')
plt.title('Keyword Distribution')
plt.show()
```

```python
sorted_keyword_counts_desc = dict(sorted(filtered_keyword_counts.items(), key=lambda item: item[1], reverse=True))
top_keywords = list(sorted_keyword_counts_desc.keys())[:5]

top_keywords
```

```python
search_string = " OR ".join([f'"{kw.strip()}"' for kw in top_keywords])
print("Generated search string:")
print(search_string)
```
