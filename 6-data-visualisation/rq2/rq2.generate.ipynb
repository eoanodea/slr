## Pie Charts - RQ2.1 Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
from matplotlib.backends.backend_pdf import PdfPages

# Create output directory if it doesn't exist
output_dir = 'output'
os.makedirs(output_dir, exist_ok=True)

# Fixed parameters for consistent sizing
RADIUS = 1.1
FIGSIZE = (6, 5.5)  # Reduced height for tighter layout
DONUT_WIDTH = 0.4
AXIS_LIMITS = 1.3
FIG_LABEL_FONT_SIZE = 12
LEGEND_FONT_SIZE = 13

# Store figures for combined PDF
figures = []

# Chart 1: Measurement Level
data1 = {
    'Label': ['HW', 'SH', 'CT', 'RQ', 'ND', 'OT'],  # Short labels for segments
    'Category': ['[HW] Hardware', '[SH] ServerHost', '[CT] Container', '[RQ] Request', '[ND] Node', '[OT] Other'],  # Full names for legend
    'Count': [18, 11, 6, 5, 4, 10],
    'Percentage': [31.03, 18.97, 10.34, 8.62, 6.90, 17.24]
}
df1 = pd.DataFrame(data1)
colors1 = ['#6baed6', '#9ecae1', '#c6dbef', '#deebf7', '#e7f4f9', '#e0e0e0']

fig1, ax1 = plt.subplots(figsize=FIGSIZE)
wedges1, texts1, autotexts1 = ax1.pie(
    df1['Percentage'], 
    labels=df1['Label'],
    autopct='%1.1f%%',
    radius=RADIUS,
    startangle=90,
    colors=colors1,
    textprops={'fontsize': FIG_LABEL_FONT_SIZE, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
ax1.axis('equal')
ax1.set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
ax1.set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
ax1.legend(wedges1, df1['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
          ncol=2, fontsize=LEGEND_FONT_SIZE, frameon=False)
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.30)
file_name = 'rq2.1.p1_pie'
fig1.savefig(os.path.join(output_dir, file_name + '.pdf'))
figures.append(fig1)
plt.show()
print(f"Chart 1 saved to: {output_dir}/{file_name}.pdf")

# Chart 2: Measurement Technique
data2 = {
    'Label': ['MB', 'DP', 'HC', 'SP', 'PM'],
    'Category': ['[MB] Model-Based', '[DP] Direct Power', '[HC] HW Counter', '[SP] SW Profiling', '[PM] Power Meter'],
    'Count': [24, 7, 4, 4, 2],
    'Percentage': [57.14, 16.67, 9.52, 9.52, 4.76]
}
df2 = pd.DataFrame(data2)
colors2 = ['#6baed6', '#9ecae1', '#c6dbef', '#deebf7', '#e7f4f9']

fig2, ax2 = plt.subplots(figsize=FIGSIZE)
wedges2, texts2, autotexts2 = ax2.pie(
    df2['Percentage'],
    labels=df2['Label'],
    autopct='%1.1f%%',
    radius=RADIUS,
    startangle=90,
    colors=colors2,
    textprops={'fontsize': FIG_LABEL_FONT_SIZE, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
ax2.axis('equal')
ax2.set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
ax2.set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
ax2.legend(wedges2, df2['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
          ncol=2, fontsize=LEGEND_FONT_SIZE, frameon=False)
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.30)
file_name = 'rq2.1.p2_pie'
fig2.savefig(os.path.join(output_dir, file_name + '.png'), dpi=300)
fig2.savefig(os.path.join(output_dir, file_name + '.pdf'))
figures.append(fig2)
plt.show()
print(f"Chart 2 saved to: {output_dir}/{file_name}.pdf")

# Chart 3: Data Collection
data3 = {
    'Label': ['MN', 'SP', 'SM', 'TR', 'OT'],
    'Category': ['[MN] Monitoring', '[SP] Sampling', '[SM] Simulation', '[TR] Tracing', '[OT] Other'],
    'Count': [18, 10, 4, 3, 4],
    'Percentage': [45.00, 25.00, 10.00, 7.50, 10.00]
}
df3 = pd.DataFrame(data3)
colors3 = ['#6baed6', '#9ecae1', '#c6dbef', '#deebf7', '#e0e0e0']

fig3, ax3 = plt.subplots(figsize=FIGSIZE)
wedges3, texts3, autotexts3 = ax3.pie(
    df3['Percentage'],
    labels=df3['Label'],
    autopct='%1.1f%%',
    radius=RADIUS,
    startangle=90,
    colors=colors3,
    textprops={'fontsize': FIG_LABEL_FONT_SIZE, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
ax3.axis('equal')
ax3.set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
ax3.set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
ax3.legend(wedges3, df3['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
          ncol=2, fontsize=LEGEND_FONT_SIZE, frameon=False)
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.30)
file_name = 'rq2.1.p3_pie'
fig3.savefig(os.path.join(output_dir, file_name + '.png'), dpi=300)
fig3.savefig(os.path.join(output_dir, file_name + '.pdf'))
figures.append(fig3)
plt.show()
print(f"Chart 3 saved to: {output_dir}/{file_name}.png and .pdf")

# Create combined horizontal figure with all three charts
print("\nCreating combined horizontal figure...")
fig_combined, axes = plt.subplots(1, 3, figsize=(18, 6))

# Chart 1 in combined figure
wedges1_c, texts1_c, autotexts1_c = axes[0].pie(
    df1['Percentage'], 
    labels=df1['Label'],
    autopct='%1.1f%%',
    radius=RADIUS,
    startangle=90,
    colors=colors1,
    textprops={'fontsize': FIG_LABEL_FONT_SIZE, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
axes[0].axis('equal')
axes[0].set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
axes[0].set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
axes[0].legend(wedges1_c, df1['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
              ncol=2, fontsize=LEGEND_FONT_SIZE, frameon=False)
axes[0].set_title('Measurement Level', fontsize=FIG_LABEL_FONT_SIZE, fontweight='bold', pad=10)

# Chart 2 in combined figure
wedges2_c, texts2_c, autotexts2_c = axes[1].pie(
    df2['Percentage'],
    labels=df2['Label'],
    autopct='%1.1f%%',
    radius=RADIUS,
    startangle=90,
    colors=colors2,
    textprops={'fontsize': 11, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
axes[1].axis('equal')
axes[1].set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
axes[1].set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
axes[1].legend(wedges2_c, df2['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
              ncol=2, fontsize=LEGEND_FONT_SIZE, frameon=False)
axes[1].set_title('Measurement Technique', fontsize=FIG_LABEL_FONT_SIZE, fontweight='bold', pad=10)

# Chart 3 in combined figure
wedges3_c, texts3_c, autotexts3_c = axes[2].pie(
    df3['Percentage'],
    labels=df3['Label'],
    autopct='%1.1f%%',
    radius=RADIUS,
    startangle=90,
    colors=colors3,
    textprops={'fontsize': 11, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
axes[2].axis('equal')
axes[2].set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
axes[2].set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
axes[2].legend(wedges3_c, df3['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
              ncol=2, fontsize=LEGEND_FONT_SIZE, frameon=False)
axes[2].set_title('Data Collection', fontsize=FIG_LABEL_FONT_SIZE, fontweight='bold', pad=10)

plt.tight_layout()

# Save combined figure
combined_pdf_path = os.path.join(output_dir, 'rq2.1_all_charts_combined.pdf')
combined_png_path = os.path.join(output_dir, 'rq2.1_all_charts_combined.png')
fig_combined.savefig(combined_pdf_path, dpi=300, bbox_inches='tight')
plt.show()
        
# Close all figures
for fig in figures:
    plt.close(fig)
plt.close(fig_combined)

print(f"✓ Combined horizontal figure saved to: {combined_pdf_path}")
print(f"✓ Combined horizontal figure saved to:")
# Print breakdown including abbreviation meanings
print("\n=== Chart Details ===\n")
print("Chart 1 - Measurement Level:")
print("  HW  = Hardware: 18 (31.03%)")
print("  SH  = ServerHost: 11 (18.97%)")
print("  CT  = Container (ContainerPod): 6 (10.34%)")
print("  RQ  = Request: 5 (8.62%)")
print("  ND  = Node (NodeDevice): 4 (6.90%)")
print("  OT  = Other: 10 (17.24%)")
print("        [Application(2), ClusterDatacenter(2), Component(2),")
print("         Network(2), OS(2), Service(2)]\n")

print("Chart 2 - Measurement Technique:")
print("  MB = Model-Based (ModelBasedEstimation): 24 (57.14%)")
print("  DP = Direct Power (DirectPowerMeasurement): 7 (16.67%)")
print("  HC = HW Counter (HardwareCounterAPI): 4 (9.52%)")
print("  SP = SW Profiling (SoftwareProfiling): 4 (9.52%)")
print("  PM = Power Meter (PhysicalPowerMeter): 2 (4.76%)\n")

print("Chart 3 - Data Collection:")
print("  MN = Monitoring: 18 (45.00%)")
print("  SP = Sampling: 10 (25.00%)")
print("  SM = Simulation: 4 (10.00%)")
print("  TR = Tracing (TracingProfiling): 3 (7.50%)")
print("  OT = Other: 4 (10.00%)")
print("        [APICounterReadout(2), Logging(2)]")

print("\n✓ Charts use horizontal legends below donuts for easy 3-column layout")
print("✓ Works in both color and black & white printing")
```
