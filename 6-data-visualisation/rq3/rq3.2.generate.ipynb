```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

## Nested Donut Chart - RQ3.2 Analysis

```python
# Create output directory if it doesn't exist
output_dir = 'output'
os.makedirs(output_dir, exist_ok=True)

# Combined nested donut chart showing all four parameters
print("Creating combined nested donut chart for RQ3.2...")

# Prepare data for nested donut
# Inner ring: four main categories
inner_labels = ['Quality\nAttribute', 'Trade-Off\nRelationship', 'Trade-Off\nSeverity', 'Trade-Off\nStrategy']
inner_sizes = [62, 40, 25, 48]  # Total counts for each category
inner_colors = ['#3182bd', '#9e9ac8', '#74c476', '#fd8d3c']

# Outer ring: Items with >=3 occurrences + Other (simplified for academic paper)
outer_labels = (
    # QualityAttribute (5 items >= 3 + other)
    ['Performance', 'QoS/SLA', 'Reliability/Avail.', 'Cost', 'Resource Util.', 'Other (7 items)'] +
    # TradeOffRelationship (4 items >= 2 + other)
    ['Balancing', 'Degrading', 'Improving', 'Inverse', 'Other (2 items)'] +
    # TradeOffSeverity (3 items >= 4 + other)
    ['Moderate', 'Severe', 'Minor', 'Other (1 item)'] +
    # TradeOffStrategy (7 items >= 3 + other)
    ['Scheduling', 'Adaptive Policy', 'Autoscale/QoS', 'Cold Start Mitig.', 
     'Multi-Obj. Optim.', 'Place/Coloc/Cache', 'Reconfig/Redist.', 'Other (6 items)']
)

outer_sizes = (
    # QualityAttribute: 5 items >= 3, other = 8 items (4×2 + 4×1)
    [31, 10, 5, 3, 3, 10] +
    # TradeOffRelationship: 4 items >= 2, other = 2 items (1×1 + 1×1)
    [21, 13, 2, 2, 2] +
    # TradeOffSeverity: 3 items >= 4, other = 1 item (1×1)
    [14, 6, 4, 1] +
    # TradeOffStrategy: 7 items >= 3, other = 6 items (3×2 + 3×1)
    [11, 10, 6, 3, 3, 3, 3, 9]
)

# Colors for outer ring - gradient within each category
outer_colors = (
    # QualityAttribute - blue gradient (6 colors, darker to lighter)
    ['#08306b', '#08519c', '#2171b5', '#4292c6', '#6baed6', '#e0e0e0'] +
    # TradeOffRelationship - purple gradient (5 colors, darker to lighter)
    ['#54278f', '#756bb1', '#9e9ac8', '#bcbddc', '#e0e0e0'] +
    # TradeOffSeverity - green gradient (4 colors, darker to lighter)
    ['#238b45', '#41ab5d', '#74c476', '#e0e0e0'] +
    # TradeOffStrategy - orange gradient (8 colors, darker to lighter)
    ['#d94801', '#f16913', '#fd8d3c', '#fdae6b', '#fdd0a2', '#fee6ce', '#fff5eb', '#e0e0e0']
)

# Create figure
fig, ax = plt.subplots(figsize=(12, 10))

# Plot outer ring (subcategories)
wedges_outer, texts_outer, autotexts_outer = ax.pie(
    outer_sizes,
    labels=outer_labels,
    autopct=lambda pct: f'{pct:.0f}%' if pct >= 2 else '',
    radius=1.3,
    colors=outer_colors,
    wedgeprops={'width': 0.4, 'edgecolor': 'white', 'linewidth': 2},
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    pctdistance=0.85,
    labeldistance=1.15,
    startangle=90
)

# Plot inner ring (main categories)
wedges_inner, texts_inner = ax.pie(
    inner_sizes,
    labels=inner_labels,
    radius=0.9,
    colors=inner_colors,
    wedgeprops={'width': 0.5, 'edgecolor': 'white', 'linewidth': 3},
    textprops={'fontsize': 16, 'fontweight': 'bold'},
    labeldistance=0.32,
    startangle=90
)

ax.axis('equal')

plt.tight_layout()

# Save the combined nested chart
nested_file_path = os.path.join(output_dir, 'rq3.2_nested_combined.pdf')
fig.savefig(nested_file_path, dpi=300, bbox_inches='tight')
plt.show()
plt.close(fig)

print(f"✓ Combined nested donut saved to: {nested_file_path}")
print(f"\nInner ring totals:")
print(f"  Quality Attribute: {inner_sizes[0]} items")
print(f"  Trade-Off Relationship: {inner_sizes[1]} items")
print(f"  Trade-Off Severity: {inner_sizes[2]} items")
print(f"  Trade-Off Strategy: {inner_sizes[3]} items")

# Output data in copy-pasteable format
print("\n" + "="*80)
print("NESTED DONUT CHART DATA - Copy/Paste Format")
print("="*80)
print("\nINNER RING (Main Categories):")
print("-" * 40)
for label, size in zip(inner_labels, inner_sizes):
    print(f"  {label.replace(chr(10), ' ')}: {size} items")

print("\n\nOUTER RING (Subcategories by Main Category):")
print("-" * 40)

print("\n1. QUALITY ATTRIBUTE (62 items total - Items ≥3 + Other):")
qa_labels = ['Performance', 'QoS/SLA', 'Reliability/Availability', 'Cost', 'Resource Utilization',
             'Other (Energy Efficiency(2), Sustainability(2), User Experience(2), Consistency(1), Output Quality(1), Scalability(1), Security(1))']
qa_sizes = [31, 10, 5, 3, 3, 10]
for label, size in zip(qa_labels, qa_sizes):
    pct = (size / 62) * 100
    print(f"   - {label}: {size} ({pct:.2f}%)")

print("\n2. TRADE-OFF RELATIONSHIP (40 items total - Items ≥2 + Other):")
tr_labels = ['Balancing', 'Degrading', 'Improving', 'Inverse',
             'Other (Explicit Trade-Off(1), Reversing(1))']
tr_sizes = [21, 13, 2, 2, 2]
for label, size in zip(tr_labels, tr_sizes):
    pct = (size / 40) * 100
    print(f"   - {label}: {size} ({pct:.2f}%)")

print("\n3. TRADE-OFF SEVERITY (25 items total - Items ≥4 + Other):")
ts_labels = ['Moderate', 'Severe', 'Minor', 'Other (Significant(1))']
ts_sizes = [14, 6, 4, 1]
for label, size in zip(ts_labels, ts_sizes):
    pct = (size / 25) * 100
    print(f"   - {label}: {size} ({pct:.2f}%)")

print("\n4. TRADE-OFF STRATEGY (48 items total - Items ≥3 + Other):")
tos_labels = ['Scheduling', 'Adaptive Policy', 'Autoscaling/QoS Control', 
              'Cold Start Mitigation', 'Multi-Objective Optimization',
              'Placement/Colocation/Caching', 'Reconfiguration/Redistribution',
              'Other (Brownout/Optionality(2), HW Accel(2), Tuning(2), Migration/Dupl(1), RL Control(1), Replication/Scaling(1))']
tos_sizes = [11, 10, 6, 3, 3, 3, 3, 9]
for label, size in zip(tos_labels, tos_sizes):
    pct = (size / 48) * 100
    print(f"   - {label}: {size} ({pct:.2f}%)")

print("\n" + "="*80)
```
