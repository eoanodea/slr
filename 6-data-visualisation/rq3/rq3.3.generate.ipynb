```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

## Nested Donut Chart - RQ3.3 Analysis

```python
# Create output directory if it doesn't exist
output_dir = 'output'
os.makedirs(output_dir, exist_ok=True)

# Combined nested donut chart showing all three parameters
print("Creating combined nested donut chart for RQ3.3...")

# Prepare data for nested donut
# Inner ring: three main categories
inner_labels = ['Challenge\nType', 'Adoption\nPhase', 'Challenge\nComplexity']
inner_sizes = [51, 35, 20]  # Total counts for each category
inner_colors = ['#3182bd', '#fd8d3c', '#74c476']

# Outer ring: Items with >=3 occurrences + Other (simplified for academic paper)
outer_labels = (
    # ChallengeType (9 items >= 3 + other)
    ['Measurement', 'Technical', 'Performance Interference', 'Predictability/Control', 'Data Availability',
     'Evaluation Validity', 'Tooling', 'Deployment/Operational', 'Resource Heterogeneity', 'Other (2 items)'] +
    # AdoptionPhase (3 items >= 5 + other)
    ['Operation', 'Implementation', 'Design', 'Other (4 items)'] +
    # ChallengeComplexity (1 item >= 3 + other)
    ['Complex', 'Other (2 items)']
)

outer_sizes = (
    # ChallengeType: 9 items >= 3, other = 2 items (1×2 + 1×1)
    [10, 9, 6, 5, 4, 4, 4, 3, 3, 3] +
    # AdoptionPhase: 3 items >= 5, other = 4 items (4×1)
    [20, 6, 5, 4] +
    # ChallengeComplexity: 1 item >= 3, other = 2 items (2×1)
    [18, 2]
)

# Colors for outer ring - gradient within each category
outer_colors = (
    # ChallengeType - blue gradient (10 colors, darker to lighter)
    ['#08306b', '#08519c', '#2171b5', '#4292c6', '#6baed6', '#9ecae1',
     '#c6dbef', '#deebf7', '#e7f4f9', '#e0e0e0'] +
    # AdoptionPhase - orange gradient (4 colors, darker to lighter)
    ['#d94801', '#f16913', '#fd8d3c', '#e0e0e0'] +
    # ChallengeComplexity - green gradient (2 colors, darker to lighter)
    ['#238b45', '#e0e0e0']
)


def is_dark_color(rgba, threshold=0.5):
    r, g, b, _ = rgba
    luminance = 0.299 * r + 0.587 * g + 0.114 * b
    return luminance < threshold


# Create figure
fig, ax = plt.subplots(figsize=(12, 10))

# Plot outer ring (subcategories)
wedges_outer, texts_outer, autotexts_outer = ax.pie(
    outer_sizes,
    labels=outer_labels,
    autopct=lambda pct: f'{pct:.0f}%' if pct >= 2 else '',
    radius=1.3,
    colors=outer_colors,
    wedgeprops={'width': 0.4, 'edgecolor': 'white', 'linewidth': 2},
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    pctdistance=0.85,
    labeldistance=1.15,
    startangle=90
)

# Make percentage text white on dark wedges, black otherwise
for wedge, autotext in zip(wedges_outer, autotexts_outer):
    if autotext.get_text().strip():
        autotext.set_color('white' if is_dark_color(wedge.get_facecolor()) else 'black')
        autotext.set_fontweight('bold')

# Plot inner ring (main categories)
wedges_inner, texts_inner = ax.pie(
    inner_sizes,
    labels=inner_labels,
    radius=0.9,
    colors=inner_colors,
    wedgeprops={'width': 0.5, 'edgecolor': 'white', 'linewidth': 3},
    textprops={'fontsize': 16, 'fontweight': 'bold'},
    labeldistance=0.32,
    startangle=90
)

ax.axis('equal')

plt.tight_layout()

# Save the combined nested chart
nested_file_path = os.path.join(output_dir, 'rq3.3_nested_combined.pdf')
fig.savefig(nested_file_path, dpi=300, bbox_inches='tight')
plt.show()
plt.close(fig)

print(f"✓ Combined nested donut saved to: {nested_file_path}")
print(f"\nInner ring totals:")
print(f"  Challenge Type: {inner_sizes[0]} items")
print(f"  Adoption Phase: {inner_sizes[1]} items")
print(f"  Challenge Complexity: {inner_sizes[2]} items")

# Output data in copy-pasteable format
print("\n" + "="*80)
print("NESTED DONUT CHART DATA - Copy/Paste Format")
print("="*80)
print("\nINNER RING (Main Categories):")
print("-" * 40)
for label, size in zip(inner_labels, inner_sizes):
    print(f"  {label.replace(chr(10), ' ')}: {size} items")

print("\n\nOUTER RING (Subcategories by Main Category):")
print("-" * 40)

print("\n1. CHALLENGE TYPE (51 items total - Items ≥3 + Other):")
ct_labels = ['Measurement', 'Technical', 'Performance Interference', 'Predictability/Control',
             'Data Availability', 'Evaluation Validity', 'Tooling', 'Deployment/Operational',
             'Resource Heterogeneity', 'Other (Organizational(2), Programmability(1))']
ct_sizes = [10, 9, 6, 5, 4, 4, 4, 3, 3, 3]
for label, size in zip(ct_labels, ct_sizes):
    pct = (size / 51) * 100
    print(f"   - {label}: {size} ({pct:.2f}%)")

print("\n2. ADOPTION PHASE (35 items total - Items ≥5 + Other):")
ap_labels = ['Operation', 'Implementation', 'Design',
             'Other (Deployment(1), Evaluation(1), Planning(1), Production(1))']
ap_sizes = [20, 6, 5, 4]
for label, size in zip(ap_labels, ap_sizes):
    pct = (size / 35) * 100
    print(f"   - {label}: {size} ({pct:.2f}%)")

print("\n3. CHALLENGE COMPLEXITY (20 items total - Items ≥3 + Other):")
cc_labels = ['Complex', 'Other (Moderate(1), Systemic(1))']
cc_sizes = [18, 2]
for label, size in zip(cc_labels, cc_sizes):
    pct = (size / 20) * 100
    print(f"   - {label}: {size} ({pct:.2f}%)")

print("\n" + "="*80)
```
