## Pie Charts - RQ3.1 Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
from matplotlib.backends.backend_pdf import PdfPages

# Create output directory if it doesn't exist
output_dir = 'output'
os.makedirs(output_dir, exist_ok=True)

# Fixed parameters for consistent sizing
RADIUS = 1.1
FIGSIZE = (6, 5.5)
DONUT_WIDTH = 0.4
AXIS_LIMITS = 1.3
FIG_LABEL_FONT_SIZE = 12
LEGEND_FONT_SIZE = 13

# Store figures for combined PDF
figures = []

# Function to display percentages only if >= 2%
def autopct_format(pct):
    return f'{pct:.1f}%' if pct >= 2 else ''

# Chart 1: Modernization Strategy
data1 = {
    'Label': ['SR', 'PM', 'BD', 'RR', 'AS', 'OP', 'PD', 'SA', 'SD', 'HR', 'OF', 'RD', 'WB', 'OT'],
    'Category': [
        '[SR] Scheduling/Resource Opt.',
        '[PM] Placement/Migration',
        '[BD] Brownout/Deactivation',
        '[RR] Refactoring/Rearchitecting',
        '[AS] Autoscaling/QoS Control',
        '[OP] Orchestration/Provisioning',
        '[PD] Power Management/DVFS',
        '[SA] Serverless Adoption',
        '[SD] Sustainability By Design',
        '[HR] Hardware Redesign',
        '[OF] Offloading',
        '[RD] Redundancy/Duplication',
        '[WB] Workload Batching/Compute Opt.',
        '[OT] Other (e.g., Comm. Reconfig.)'
    ],
    'Count': [13, 8, 5, 4, 3, 3, 3, 3, 3, 2, 2, 2, 2, 1],
    'Percentage': [24.07, 14.81, 9.26, 7.41, 5.56, 5.56, 5.56, 5.56, 5.56, 3.70, 3.70, 3.70, 3.70, 1.85]
}
df1 = pd.DataFrame(data1)
colors1 = ['#08519c', '#2171b5', '#4292c6', '#6baed6', '#9ecae1', '#c6dbef', '#deebf7',
           '#e7f4f9', '#f0f0f0', '#e0e0e0', '#d0d0d0', '#c0c0c0', '#b0b0b0', '#e0e0e0']

fig1, ax1 = plt.subplots(figsize=FIGSIZE)
wedges1, texts1, autotexts1 = ax1.pie(
    df1['Percentage'], 
    labels=df1['Label'],
    autopct=autopct_format,
    radius=RADIUS,
    startangle=90,
    colors=colors1,
    textprops={'fontsize': 10, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
ax1.axis('equal')
ax1.set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
ax1.set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
ax1.legend(wedges1, df1['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
          ncol=2, fontsize=11, frameon=False)
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.35)
file_name = 'rq3.1.p1_pie'
fig1.savefig(os.path.join(output_dir, file_name + '.pdf'))
figures.append(fig1)
plt.show()
print(f"Chart 1 saved to: {output_dir}/{file_name}.pdf")

# Chart 2: Solution Category
data2 = {
    'Label': ['SC', 'AS', 'WM', 'PD', 'OR', 'PM', 'RR', 'OT'],
    'Category': [
        '[SC] Scheduling',
        '[AS] Autoscaling',
        '[WM] Workload Management',
        '[PD] Placement/Deployment',
        '[OR] Orchestration',
        '[PM] Power Management',
        '[RR] Redundancy/Replication',
        '[OT] Other (e.g., Code Opt., Security)'
    ],
    'Count': [16, 10, 8, 7, 5, 2, 2, 4],
    'Percentage': [29.63, 18.52, 14.81, 12.96, 9.26, 3.70, 3.70, 7.40]
}
df2 = pd.DataFrame(data2)
colors2 = ['#08519c', '#2171b5', '#4292c6', '#6baed6', '#9ecae1', '#c6dbef', 
           '#deebf7', '#e0e0e0']

fig2, ax2 = plt.subplots(figsize=FIGSIZE)
wedges2, texts2, autotexts2 = ax2.pie(
    df2['Percentage'],
    labels=df2['Label'],
    autopct=autopct_format,
    radius=RADIUS,
    startangle=90,
    colors=colors2,
    textprops={'fontsize': FIG_LABEL_FONT_SIZE, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
ax2.axis('equal')
ax2.set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
ax2.set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
ax2.legend(wedges2, df2['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
          ncol=2, fontsize=LEGEND_FONT_SIZE, frameon=False)
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.30)
file_name = 'rq3.1.p2_pie'
fig2.savefig(os.path.join(output_dir, file_name + '.png'), dpi=300)
fig2.savefig(os.path.join(output_dir, file_name + '.pdf'))
figures.append(fig2)
plt.show()
print(f"Chart 2 saved to: {output_dir}/{file_name}.pdf")

# Chart 3: Implementation Technologies
data3 = {
    'Label': ['CR', 'OP', 'SIM', 'OS', 'CP', 'HP', 'NT', 'SP', 'DS', 'EI', 'ML', 'OT'],
    'Category': [
        '[CR] Container Runtime',
        '[OP] Orchestration Platform',
        '[SIM] Simulator',
        '[OS] Other Software',
        '[CP] Cloud Platform',
        '[HP] Hardware Platform',
        '[NT] Networking Tech',
        '[SP] Serverless Platform',
        '[DS] Data Store',
        '[EI] Edge/IoT Framework',
        '[ML] ML Framework',
        '[OT] Other (e.g., Optim., Service Mesh)'
    ],
    'Count': [15, 12, 5, 4, 3, 3, 3, 3, 2, 2, 2, 2],
    'Percentage': [26.79, 21.43, 8.93, 7.14, 5.36, 5.36, 5.36, 5.36, 3.57, 3.57, 3.57, 3.57]
}
df3 = pd.DataFrame(data3)
colors3 = ['#08519c', '#2171b5', '#4292c6', '#6baed6', '#9ecae1', '#c6dbef', '#deebf7',
           '#e7f4f9', '#f0f0f0', '#e0e0e0', '#d0d0d0', '#e0e0e0']

fig3, ax3 = plt.subplots(figsize=FIGSIZE)
wedges3, texts3, autotexts3 = ax3.pie(
    df3['Percentage'],
    labels=df3['Label'],
    autopct=autopct_format,
    radius=RADIUS,
    startangle=90,
    colors=colors3,
    textprops={'fontsize': 10, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
ax3.axis('equal')
ax3.set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
ax3.set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
ax3.legend(wedges3, df3['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
          ncol=2, fontsize=11, frameon=False)
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.35)
file_name = 'rq3.1.p3_pie'
fig3.savefig(os.path.join(output_dir, file_name + '.png'), dpi=300)
fig3.savefig(os.path.join(output_dir, file_name + '.pdf'))
figures.append(fig3)
plt.show()
print(f"Chart 3 saved to: {output_dir}/{file_name}.png and .pdf")

# Create combined horizontal figure with all three charts
print("\nCreating combined horizontal figure...")
fig_combined, axes = plt.subplots(1, 3, figsize=(18, 6))

# Chart 1 in combined figure
wedges1_c, texts1_c, autotexts1_c = axes[0].pie(
    df1['Percentage'], 
    labels=df1['Label'],
    autopct=autopct_format,
    radius=RADIUS,
    startangle=90,
    colors=colors1,
    textprops={'fontsize': 10, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
axes[0].axis('equal')
axes[0].set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
axes[0].set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
axes[0].legend(wedges1_c, df1['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
              ncol=2, fontsize=11, frameon=False)
axes[0].set_title('Modernization Strategy', fontsize=FIG_LABEL_FONT_SIZE, fontweight='bold', pad=10)

# Chart 2 in combined figure
wedges2_c, texts2_c, autotexts2_c = axes[1].pie(
    df2['Percentage'],
    labels=df2['Label'],
    autopct=autopct_format,
    radius=RADIUS,
    startangle=90,
    colors=colors2,
    textprops={'fontsize': 11, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
axes[1].axis('equal')
axes[1].set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
axes[1].set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
axes[1].legend(wedges2_c, df2['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
              ncol=2, fontsize=LEGEND_FONT_SIZE, frameon=False)
axes[1].set_title('Solution Category', fontsize=FIG_LABEL_FONT_SIZE, fontweight='bold', pad=10)

# Chart 3 in combined figure
wedges3_c, texts3_c, autotexts3_c = axes[2].pie(
    df3['Percentage'],
    labels=df3['Label'],
    autopct=autopct_format,
    radius=RADIUS,
    startangle=90,
    colors=colors3,
    textprops={'fontsize': 10, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'width': DONUT_WIDTH},
    pctdistance=0.85,
    labeldistance=1.05
)
axes[2].axis('equal')
axes[2].set_xlim(-AXIS_LIMITS, AXIS_LIMITS)
axes[2].set_ylim(-AXIS_LIMITS, AXIS_LIMITS)
axes[2].legend(wedges3_c, df3['Category'], loc='upper center', bbox_to_anchor=(0.5, -0.02), 
              ncol=2, fontsize=11, frameon=False)
axes[2].set_title('Implementation Technologies', fontsize=FIG_LABEL_FONT_SIZE, fontweight='bold', pad=10)

plt.tight_layout()

# Save combined figure
combined_pdf_path = os.path.join(output_dir, 'rq3.1_all_charts_combined.pdf')
combined_png_path = os.path.join(output_dir, 'rq3.1_all_charts_combined.png')
fig_combined.savefig(combined_pdf_path, dpi=300, bbox_inches='tight')
plt.show()
        
# Close all figures
for fig in figures:
    plt.close(fig)
plt.close(fig_combined)

print(f"✓ Combined horizontal figure saved to: {combined_pdf_path}")

# Print breakdown
print("\n=== Chart Details ===\n")
print("Chart 1 - Modernization Strategy:")
for idx, row in df1.iterrows():
    print(f"  {row['Label']:3s} = {row['Category']:45s} {row['Count']:2d} ({row['Percentage']:5.2f}%)")

print("\nChart 2 - Solution Category:")
for idx, row in df2.iterrows():
    print(f"  {row['Label']:3s} = {row['Category']:45s} {row['Count']:2d} ({row['Percentage']:5.2f}%)")

print("\nChart 3 - Implementation Technologies:")
for idx, row in df3.iterrows():
    print(f"  {row['Label']:3s} = {row['Category']:45s} {row['Count']:2d} ({row['Percentage']:5.2f}%)")

print("\n✓ Charts use horizontal legends below donuts for easy 3-column layout")
print("✓ Works in both color and black & white printing")
```
